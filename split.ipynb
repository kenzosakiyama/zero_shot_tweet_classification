# Split

O objetivo deste notebook é a geração de conjuntos de treino e teste para validação cruzada baseada em *holdout* em uma tarefa de classificação de texto (análise de sentimentos). 

```python
import pandas as pd
from sklearn.model_selection import train_test_split
import seaborn as sns
```

# Carregando os dados

```python
df = pd.read_csv("tweetsentbr.csv", index_col=0)
```

# Criando splits de treino e teste

Os splits serão criados de forma estratificada, visando manter a distribuição de classes em nos conjuntos de treino e teste.

```python
test_pct = 0.2
seed = 1234
```

```python
train_df, test_df = train_test_split(df, test_size=test_pct, shuffle=True, random_state=seed, stratify=df["label"])
```

```python
len(train_df), len(test_df)
```

# Visualizando distribuição de classes em cada split

- 0: tweets neutros
- 1: tweets positivos
- 2: tweets negativos

```python
sns.countplot(data=train_df, x=train_df["label"])
```

```python
sns.countplot(data=test_df, x=test_df["label"])
```

```python
train_df.to_csv("tweetsentbr_train.csv")
test_df.to_csv("tweetsentbr_test.csv")
```
